```python
#hypothesis testing:
import numpy as np
from scipy import stats


```

```python
#two grps of data
data1=np.array([28,21,26,29,23])
data2=np.array([21,27,25,28,19])
print(data1.mean())
print(data2.mean())
```

```python
#perform t-test assuming unequal variances
t_stat,p_value=stats.ttest_ind(data1,data2,equal_var=False)
```

```python
print("T-statistics:",t_stat)
print("P-value:",p_value)
```

```python
alpha=0.05
```

```python
#interrupt the results
if p_value < alpha:
  print("Reject the null hypothesis.The means are significant.")
else:
  print("Fail to reject the null hypothesis.The means are not significant")
```

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Vaccine.csv")
df
```

```python
a=df['Efficiency']
print("Mean efficiency:",a)
b=df["Rec_Rate"].mean()
print("mean Rc_Rate,",b)
data1=df['Efficiency']
data2=df['Rec_Rate']
t_stat ,p_value=stats.ttest_ind(data1,data2,equal_var=False)
print("T-statistic:",t_stat)
print("P_value:",p_value)
```

```python
alpha=0.05
```

```python
if p_value<alpha:
  print("H0-related")
else:
  print("H1-not related")
```

# ANOVA **TESTING**

```python
dfA=df[df['Vaccine']=='A']['Efficiency']
dfB=df[df['Vaccine']=='B']['Efficiency']
dfA
```

```python
f_stat,p_value=stats.f_oneway(dfA,dfB)
print("F-statistic:",f_stat)
print("p-value:",p_value)
```

```python
alpha=0.05
if p_value>alpha:
  print("H0-related")
else:
  print("H1-not related")
```
