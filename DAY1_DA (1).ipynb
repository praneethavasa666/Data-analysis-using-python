```python
#take a number from the user which is a salary.If the salary is more than 50000,there is a 10% tax deduction.Print his final salary.
s=int(input("Enter the salary"))
if(s>=50000):
  s=s-(s*(10/100))
  print("Salary is",s)
else:
  print("Salary is less than 50000")
```

```python
#get the salary from the user.If the salary is more than or equal to 50k.apply the 10% tax deduction.if the salary is less than 50k,apply 5% tax.If the salary is more than 70k,apply 12% tax.If the salary exceeds 1lakh print you are promoted to nxt level.
s=int(input("Enter the salary"))
if(s<=50000):
  s=s-(s*(5/100))
  print("Salary is",s)
elif(s>=50000 and s<70000):
  s=s-(s*(10/100))
  print("Salary is",s)
elif(s>=70000 and s<100000):
  s=s-(s*(12/100))
  print("Salary is",s)
else:
  print("You are promoted")
```

```python
#list comprehension
#new list=[output for]
s=[23,12,64,87,32]
p=[i*i for i in s]
print(p)
```

```python
p=[i*i for i in s if i>50]
print(p)
```

```python
#list comprehension
#if-else
#new list=[(st of if) if(cond) else (st of else) for]
l=[23,67,12,34,69]
res=[i*i if i>50 else 'X' for i in l]
print(res)
```
