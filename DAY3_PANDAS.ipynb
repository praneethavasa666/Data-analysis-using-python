**ACCESSING** **THE** **DATA**

```python
#iloc= integer  location,index
#loc=fieldnames ,index
#df.name.loc[index]--->rows
#df.name[st:stop]--->range of rows
#dfname.loc[r_ind,col_name]---->rows&col
mydf.loc[0] #first record
```

```python
mydf.loc[8:14]  #range of records
```

```python
mydf.loc[5:7,"SEM1"]  #first record
```

```python
mydf.iloc[10:12,0:2]  #both rows and indexes
```

```python
mydf[mydf.Grade == "A"]
```

```python
#print the records who scored 9.3 in sem3
mydf[mydf.SEM3==9.3]
```

```python
mydf.loc[mydf['SEM1']>9.5 , ['Names']]  #printing the name of students who scored >9.5
                                        #loc[ind,col]
```

```python
#to access the grades of students who scored more than 9 in sem3
mydf.loc[mydf.SEM3>9.7,"Grade"]
```

```python
#accessing using index fun
print(mydf[mydf.index==3])
```

```python
print(mydf.loc[mydf.index==7,"Grade"])  #using loc only we can access both rows and columns
```

```python
#removing the duplicate records
mydf.drop_duplicates(inplace=True)
print(mydf)
```
