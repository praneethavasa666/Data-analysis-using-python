```python
#There are 5 sub .psychology,nanotechnology,medicine,anchology,anthrology,2 students have applied to do masters in any of these 5 streams witha line plot show in which field they can excel.the cutoff for each sub is 70.
#s1=[67,45,73,71,57] ,s2=[43,67,68,53,85]

x=['psychology','nanotechnology','medicine','anchology','anthrology']
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cut=[70,70,70,70,70]
#std1
#syntax=plt.plottype(x,y,label,color,marker,linewidth,linestyle)
plt.plot(x,s1,label="Student1",color="c",marker="H")
#std2
plt.scatter(x,s2,label="Student2",color="m")
#cutoff
plt.plot(x,cut,label="cutoff",color="r",linestyle=":",linewidth=4)
plt.title("Student marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
```

```python
#india vs australia match is on live.India has given 298 as target for australia.the runs for every 5 overs is given as follows.[25,51,84,131,160,189,220,250,267,297],aus have completed only 25 overs.score=[15,41,94,110,151].plot overs vs runs for both the countries with india as blue line and australia as yellow line.give appropriate labels and legend.the marker for india is flag,and australia is a plus.Take overs from arange function.create runs as arrays.line width =2
import numpy as np
India=np.array([25,51,84,131,160,189,220,250,267,297])
Australia=np.array([15,41,94,110,151])
ofi=np.arange(5,51,5)
ofa=np.arange(5,26,5)
plt.title("Ind vs Aus")
plt.plot(India,ofi,label="India",color="blue",marker=">",linewidth=2)
plt.plot(Australia,ofa,label="Australia",color="yellow",marker="P")

plt.xlabel("overs")
plt.ylabel("score")
plt.legend()
plt.grid()
```
